## Booking Classification (Binary with Keras)

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

## Read the train dataset

```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
df.head()
```

## Perfrom Basic data quanlity checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

## Seperate X and Y (Booking_status)

```python
df['booking_status'].value_counts()
```

```python
df['booking_status'].value_counts().plot(kind='bar')
```

```python
X=df.drop(columns=['id','booking_status'])
Y=df[['booking_status']]
```

```python
X.head()
```

```python
Y.head()
```

#### Create a preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=([('impute',SimpleImputer(strategy='median')),
                       ('scaler',StandardScaler())])).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(X_pre,Y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.shape[1]
```

## Build a keras neural network

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

image.png

image.png

```python
model = Sequential()

# Add input layer
model.add(Input(shape=(xtrain.shape[1],)))
# Add Hidden Layer 1
model.add(Dense(units=64,activation='relu'))
# Add Dropout layer 1
model.add(Dropout(rate=0.3))
# Add Hidden layer 2
model.add(Dense(units=32,activation='relu'))
# Add Dropout layer 2
model.add(Dropout(rate=0.2))
# Output Layer
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

## Compile the model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

## Fitting the model
image.png

```python
from keras.callbacks import EarlyStopping
```

```python
es= EarlyStopping(monitor='val_loss',patience=10)
```

```python
hist=model.fit(xtrain,ytrain,validation_split=0.2,
               epochs=200,callbacks=[es])
```

### Check the learning curve for the neural network

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'],label='Train Loss')
plt.plot(hist.history['val_loss'],label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary cross Entropy Loss')
plt.title('Learnings Curve')
plt.legend()
plt.show()
```

### Evalute Model In train and Test

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

### Get the probability from neural network

```python
yprob_tr = model.predict(xtrain)
yprob_ts = model.predict(xtest)
```

## Plot the confusion matrix for test

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

### Print the classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```
